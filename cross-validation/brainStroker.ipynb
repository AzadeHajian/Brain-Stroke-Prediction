```python

l1=LabelEncoder()
data["Residence_type"]=l1.fit_transform(data["Residence_type"])
data["smoking_status"]=l1.fit_transform(data["smoking_status"])
data["ever_married"]=l1.fit_transform(data["ever_married"])
data["work_type"]=l1.fit_transform(data["work_type"])
data["gender"]=l1.fit_transform(data["gender"])
data.describe()
```

```python
data.head(10)
```

```python
X = data.drop('stroke', axis=1)  # Drop the target column ('stroke')
y = data['stroke']  # Target variable
```

```python
# Handle class imbalance using SMOTE
smote = SMOTE(random_state=42)
X_train_resampled,y_train_resampled=smote.fit_resample(X,y)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt


# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_train_resampled, y_train_resampled, test_size=0.3, random_state=42)


# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Models
models = {
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced'),
    "Random Forest": RandomForestClassifier(n_estimators=100, criterion="entropy", random_state=42),
    "Decision Tree": DecisionTreeClassifier(criterion='gini', random_state=42)
}

# Initialize a list to store the cross-validation results for each model
cv_results = []

# Cross-validation setup
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Evaluate each model
for name, model in models.items():
    # Perform cross-validation
    cv_scores = cross_val_score(model, X_train, y_train, cv=kfold, scoring='accuracy')
    cv_mean_score = cv_scores.mean()
    
    # Train the model on the entire training set
    model.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Store the results
    cv_results.append({
        "Model": name,
        "Cross-Validation Score": cv_mean_score,
        "Test Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1
    })

# Convert results to DataFrame for easy viewing
cv_results_df = pd.DataFrame(cv_results)

# Print results
print(cv_results_df)

# Plot the comparison of cross-validation scores
plt.figure(figsize=(10, 8))
plt.plot(cv_results_df["Model"], cv_results_df["Cross-Validation Score"], marker='o', linestyle='-', color='b', label='Cross-Validation Score')
plt.plot(cv_results_df["Model"], cv_results_df["Test Accuracy"], marker='s', linestyle='--', color='r', label='Test Accuracy')
plt.xlabel("Model")
plt.ylabel("Score")
plt.title("Comparison of Model Performance")
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig('cv_comparison_plot.png', dpi=300)
plt.show()
```
